```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def phi(x):
    return (1 - x**2)**(1/2)

# true integral value
I = np.pi/4

n_max = 10000

U = np.random.uniform(0, 1, n_max)
I_est = np.zeros(n_max - 1)
I_var = np.zeros(n_max - 1)

fig = plt.figure(figsize=(10 , 5))
k = 0
K = np.array([])

for n in range(1, n_max, 5):
    I_est[k] = 1/n * sum(phi(U[0:n]))
    I_var[k] = 1/(n**2) * sum((phi(U[0:n]) - I_est[k])**2)

    k = k + 1
    K = np.append(K, n)

    if (n - 1) % 200 == 0:  # why only those mod 200
        plt.cla()  # 
        plt.plot(K, I_est[0:k], 'k-', label='MC estimate')
        plt.plot(K, I_est[0:k] + np.sqrt(I_var[0:k]), 'r', label="$\sigma$", alpha=1)
        plt.plot(K, I_est[0:k] - np.sqrt(I_var[0:k]), 'r', alpha=1)
        plt.plot([0, n_max], [I, I], 'b--', label='True Value', alpha=1, linewidth=2)
        plt.legend()
        plt.xlabel('Number of samples')
        plt.ylabel('Estimate')
        plt.xlim([0, n_max])
        #
        plt.show(block=False) # For animation
        #
        plt.pause(0.0001)
```

```python
fig, axs = plt.subplots(2, 1, figsize=(10, 5))
axs[0].plot([0, n_max], [I, I], 'r-')
axs[1].plot([0, n_max, n_max], [I, I, I], 'b--')
plt.close()  # window closed

# no plot shown
axs[1].plot([0, n_max, n_max, n_max], [I, I, I, I], 'b--')
```
